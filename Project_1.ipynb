```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_excel("Superstore_USA.xlsx")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean(),inplace = True)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

# Order Priority

```python
dataset['Order Priority'].value_counts()
```

```python
dataset['Order Priority'].unique()
```

```python
dataset['Order Priority'] = dataset['Order Priority'].replace("Critical ","Critical")
```

```python
dataset['Order Priority'].value_counts()
```

```python
plt.figure(figsize = (5,4))
sns.countplot(data = dataset,x = 'Order Priority')
plt.title("Count of Order Priority",fontdict = {'fontsize':15,'color':'b','fontweight':'bold'})
plt.savefig(fname = "Count of Order Priority.jpg",dpi = 800,bbox_inches = 'tight')
plt.show()
```

# Ship Mode

```python
dataset['Ship Mode'].value_counts()
```

```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values
```

```python
# plt.figure(figsize = (5,4))
plt.pie(y,labels = x,startangle = 60,autopct = "%0.2f%%")
plt.title("Pie chart of Shipping Modes",fontdict = {'fontsize':15,'color':'g','fontweight':'bold'})
plt.legend(bbox_to_anchor = (1,1))
plt.savefig("Pie chart of Shipping modes.jpg",dpi = 800,bbox_inches = 'tight')
plt.show()
```

```python
sns.countplot(data = dataset,x = 'Ship Mode',hue = 'Product Category')
plt.title("Shipping modes & Product categories",fontdict = {'fontsize':15,'color':'b','fontweight':'bold'})
plt.xlabel("Ship Mode",fontdict = {'fontsize':15,'color':'m'})
plt.ylabel("Count",fontdict = {'fontsize':15,'color':'m'})
plt.savefig("Shipping modes & Product categories.jpg",dpi = 800,bbox_inches = 'tight')
plt.show()
```
